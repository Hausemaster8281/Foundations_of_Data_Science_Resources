```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Define hashtags
user1 = {"#fitness", "#workout", "#gym", "#health", "#training"}
user2 = {"#fitness", "#yoga", "#health", "#meditation", "#wellness"}

# Plot Venn diagram
plt.figure(figsize=(8, 5))
venn2([user1, user2], set_labels=("User 1", "User 2"))
plt.title("Hashtag Overlap (Jaccard Similarity = 0.33)")
plt.show()
```
