```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

data = np.array([8, 11, 21, 29, 36]).reshape(-1, 1)  # Reshape for 1D clustering
```

```python
Z_single = linkage(data, method='single')
plt.figure(figsize=(10, 5))
dendrogram(Z_single, labels=["8", "11", "21", "29", "36"])
plt.title("Dendrogram (Single Linkage)")
plt.show()
```

```python
Z_complete = linkage(data, method='complete')
plt.figure(figsize=(10, 5))
dendrogram(Z_complete, labels=["8", "11", "21", "29", "36"])
plt.title("Dendrogram (Complete Linkage)")
plt.show()
```

```python
Z_complete = linkage(data, method='average')
plt.figure(figsize=(10, 5))
dendrogram(Z_complete, labels=["8", "11", "21", "29", "36"])
plt.title("Dendrogram (Average Linkage)")
plt.show()
```
